## Semantic chunking

There is problem with normal chunking and that is , there are some information that might get skipped and better chunks->better retrieval->better output.
So, instead of chunking using recursive character text splitter what we do is, we do semantic chunking and because of this the context becomes more enriched with the information.
So, it is done in the following steps :

1.Document segmentation - Here, we decide how to make chunks i.e. based on full stop, next lines, comma etc
2.Sentence Embeddings
3.Semantic similarity check (using cosine similarities)
4.Merging of sentences based on threshold value
5.Finally got the merged chunk

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')

text="""
Langchain is a framework for buliding applications using LLMs.
Langchain provides modular abstractions to combine LLMs with tools like OpenAI and Pinecone.
You can create chains,memory , agents and retrievers.
The Eiffel tower is located in Paris.
France is a Popular tourist destination.


"""

sentences=[s.strip() for s in text.split("\n") if s.strip()]
embeddings=model.encode(sentences)

threshold=0.7
chunks=[]
curr_sentence=[sentences[0]]


for i in range(1,len(sentences)):
    sim=cosine_similarity([embeddings[i-1]],[embeddings[i]])[0][0]
    if sim>=threshold:
        curr_sentence.append(sentences[i])
    else:
        chunks.append(".".join(curr_sentence))
        curr_sentence=[sentences[i]]
chunks.append(". ".join(curr_sentence))  # append last chunk
print(chunks)
```
